```python
def check(word):
    n = len(word)
    for ch in word:
        if ord(ch) <= 127:
            n -= 1
    return n <= 0


def doit(title):
    str = ""
    countOfWords = 0
    words = title.split(' ')
    for word in words:
        if check(word):
            str += f"{word} "
    return str
```

```python
for link in data:
    title = doit(data[link])
    if len(title) > 0:
        data[link] = title
```

```python
def count(text):
    cnt = 0
    for city in cities:
        if text.find(city):
            cnt += 1
    return cnt

def match(text, ps):
    for p in ps:
        if p.match(text):
            return True
    return False
#     if ps[0].match(text) and count(text) >= 1:
#         return True
#     if ps[0].match(text) and count(text) >= 2:
#         return True
```

```python
import re
patterns = [r"playing\s(cricket|football|volleyball|basketball)\s?(with|in|at)?\s?(\w+)?", r"(cricket|basketball|volleyball|football)\s(match|tournament)\s(with|at|in)\s\w+"]
ps = [re.compile(p) for p in patterns]

dict = {}
for link in data:
    title = data[link]
    all = re.split('(-|:|\?|\.|\(|\||\[|\]|\s{2, 10}|#|\+|/+)', title)
    for a in all:
        if match(a.lower(), ps) and len(a) <= 50:
            dict[link] = a
```

```python
for link in dict:
    print(f"{link} - {dict[link]}")
```
